## Load and explore the data

```python
import pandas as pd
import numpy as np

# Load the datasets
company_level_metrics = pd.read_csv('./company_level_metrics.csv')
geographies = pd.read_csv('./geographies.csv')
sectors = pd.read_csv('./sectors.csv')
```

## Handling Missing Values and Data Types
Given the presence of only one missing value in ebitda_multiple_at_entry, we could opt to exclude this record from our analysis to maintain accuracy without significantly impacting our dataset's size. Furthermore, we'll convert the date columns to datetime format to facilitate temporal analyses.


```python
# Data Cleaning and Preparation

# Check for missing values
missing_values = company_level_metrics.isnull().sum()

# Check data types
data_types = company_level_metrics.dtypes

# Converting date columns to datetime
company_level_metrics['record_date'] = pd.to_datetime(company_level_metrics['record_date'], dayfirst=True)
company_level_metrics['investment_date'] = pd.to_datetime(company_level_metrics['investment_date'], dayfirst=True)

# Re-checking to ensure the cleaning was successful
missing_values_after = company_level_metrics.isnull().sum()
data_types_after = company_level_metrics[['record_date', 'investment_date']].dtypes

(missing_values_after, data_types_after)
```

## EV/EBITDA Multiple Calculation

The EV/EBITDA multiple is a financial ratio that compares a company's Enterprise Value (EV) to its Earnings Before Interest, Taxes, Depreciation, and Amortization (EBITDA). It is used to evaluate the valuation of a company. The formula for calculating the EV/EBITDA multiple at entry and at the record date is as follows:

- **EBITDA Multiple at Entry:**
$$\text{EBITDA Multiple at Entry} = \frac{\text{At Entry EV}}{\text{At Entry EBITDA}}$$

- **EBITDA Multiple at Record Date:**
$$\text{EBITDA Multiple at Record Date} = \frac{\text{At RD EV}}{\text{At RD EBITDA}}$$


```python
# Calculate the EV/EBITDA multiples
company_level_metrics['ebitda_multiple_at_entry'] = company_level_metrics['at_entry_ev'] / company_level_metrics['at_entry_ebidta']
company_level_metrics['ebitda_multiple_at_rd'] = company_level_metrics['at_rd_ev'] / company_level_metrics['at_rd_ebidta']
company_level_metrics.head()
```

```python
len(company_level_metrics)
```

```python
# Handling missing values - excluding the record with missing ebitda_multiple_at_entry
company_level_metrics = company_level_metrics.dropna(subset=['ebitda_multiple_at_entry', 'ebitda_multiple_at_rd'])
```


## Change in EV/EBITDA Multiple

To analyze the change in the valuation of the underlying companies, we calculate the difference in the EV/EBITDA multiple from the investment date to each record date. The formula for calculating this change is:

$$\text{Change in EBITDA Multiple} = \text{EBITDA Multiple at Record Date} - \text{EBITDA Multiple at Entry}$$



```python
# Calculating the change in EV/EBITDA multiples from investment to record date
company_level_metrics['change_in_ebitda_multiple'] = company_level_metrics['ebitda_multiple_at_rd'] - company_level_metrics['ebitda_multiple_at_entry']
```

```python
company_level_metrics.head()
```

### What we have done

Calculating the EV/EBITDA multiples for both entry and record dates for each company.

### what we are doing next

**Merging the company metrics with the geographies and sectors datasets to facilitate aggregation.**

Given the task involves analyzing the valuation multiples at both the entry and record dates, we will compare the entry and record date multiples directly and calculate their differences to ensure that the dataset only includes companies with complete multiple data for both the entry and record dates. It also ensures that the dataset used for analysis has complete information for both critical points, allowing for a comprehensive comparison and change calculation.


```python
# Prepare geographies and sectors for merging
# Merge the datasets
company_level_metrics = company_level_metrics.join(sectors[["sector_level_1", "sector_level_3"]].drop_duplicates().set_index("sector_level_3"), on='sector_level_3')
company_level_metrics = company_level_metrics.join(geographies[["continent", "country"]].drop_duplicates().set_index('country'), on='company_geography')

# Check the enriched dataset
company_level_metrics.head()
```


Here are a few points from the initial aggregation:
- The data has been grouped by `sector_level_1` and `continent` separately, providing us with a clear view of how the average EV/EBITDA multiples (both at entry and at record date) vary over time within each sector and continent.
- This analysis enables us to compare the valuation multiples at the fundâ€™s entry with those at the record date, across different sectors and geographical locations.


## Aggregation of Results

**Aggregating the calculated multiples by Sector Level 1 and Continent.**


For a comparative analysis across different sectors and continents, the results are aggregated by calculating the average (mean) of the EV/EBITDA multiples and their changes within each category:

- **Average EV/EBITDA Multiple by Sector or Continent:**
$$\text{Average EV/EBITDA Multiple} = \frac{\sum (\text{EBITDA Multiple for each company in the category})}{\text{Number of companies in the category}}$$

- **Average Change in EV/EBITDA Multiple by Sector or Continent:**
$$\text{Average Change in EBITDA Multiple} = \frac{\sum (\text{Change in EBITDA Multiple for each company in the category})}{\text{Number of companies in the category}}$$

These formulas provide a structured approach to analyze the valuation multiples of the underlying companies, enabling a comparative assessment of their valuations over time and across different sectors and geographies.


1. First, we will compare the **entry and record date multiples** directly and get the description of the dataframe in terms of e.g. avg and mean.

```python
# Analyzing the distribution of multiples at entry across sectors and continents
entry_distribution_by_sector = company_level_metrics.groupby('sector_level_1')['ebitda_multiple_at_entry'].describe()
entry_distribution_by_continent = company_level_metrics.groupby('continent')['ebitda_multiple_at_entry'].describe()
entry_distribution_by_sector
```

```python
# Analyzing the distribution of multiples at record date across sectors and continents
rd_distribution_by_sector = company_level_metrics.groupby('sector_level_1')['ebitda_multiple_at_rd'].describe()
rd_distribution_by_continent = company_level_metrics.groupby('continent')['ebitda_multiple_at_rd'].describe()
entry_distribution_by_sector
```

2. Next, we calculate their **differences of the multiples** between the entry and record date and get description of that difference change data.

```python
# Analyzing the distribution of change in multiples across sectors and continents
change_distribution_by_sector = company_level_metrics.groupby('sector_level_1')['change_in_ebitda_multiple'].describe()
change_distribution_by_continent = company_level_metrics.groupby('continent')['change_in_ebitda_multiple'].describe()

entry_distribution_by_sector
```

# Visualizing the data and drawing conclusions

Some pre-processing of the data before visualization.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
```

Visualizes the aggregated data as time series plots, one for sector-level trends and another for continental trends.

We'll start with time series plots for each sector and continent, followed by bar charts to compare the initial and most recent multiples more directly.

```python
# Drop rows with missing values in 'ebitda_multiple_at_entry' and 'ebitda_multiple_at_rd'
company_metrics_cleaned = company_level_metrics.dropna(subset=['ebitda_multiple_at_entry', 'ebitda_multiple_at_rd'])

# Aggregating EV/EBITDA multiples by Sector Level 1 and Continent
aggregated_by_sector = company_level_metrics.groupby(['sector_level_1', 'record_date'])[['ebitda_multiple_at_entry', 'ebitda_multiple_at_rd']].mean().reset_index()
aggregated_by_continent = company_level_metrics.groupby(['continent', 'record_date'])[['ebitda_multiple_at_entry', 'ebitda_multiple_at_rd']].mean().reset_index()

# Visualization of EV/EBITDA Multiples Over Time by Sector and Continent
plt.figure(figsize=(16, 12))

# Time series line plot for Sectors
plt.subplot(2, 1, 1)
sns.lineplot(data=aggregated_by_sector, x='record_date', y='ebitda_multiple_at_rd', hue='sector_level_1', marker='o', palette='tab10')
plt.title('EV/EBITDA Multiples Over Time by Sector', fontsize=16)
plt.xlabel('Record Date', fontsize=14)
plt.ylabel('EV/EBITDA Multiple', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Sector', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

# Time series line plot for Continents
plt.subplot(2, 1, 2)
sns.lineplot(data=aggregated_by_continent, x='record_date', y='ebitda_multiple_at_rd', hue='continent', marker='o', palette='coolwarm')
plt.title('EV/EBITDA Multiples Over Time by Continent', fontsize=16)
plt.xlabel('Record Date', fontsize=14)
plt.ylabel('EV/EBITDA Multiple', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Continent', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

plt.tight_layout()
plt.show()
```


The visualizations provide a detailed view of the EV/EBITDA multiples over time, segmented by Sector Level 1 and Continent:

- **EV/EBITDA Multiples Over Time by Sector:** This chart illustrates how the valuation multiples have evolved within different sectors. Each sector's trend line offers insights into its performance and how investors' valuation perceptions have changed over the study period.

- **EV/EBITDA Multiples Over Time by Continent:** Similarly, this chart displays the variation in multiples across continents, highlighting geographical differences in company valuations over time.

### Key Observations:

### Further Exploration and Visualization

From the previous observation and analysis, we can further explore:
- **Temporal Trends:** Analyzing how these changes vary over time, identifying any temporal patterns or correlations with economic cycles.
- **Sector and Geography Interplay:** Examining how sector trends interact with geographical variations, potentially uncovering unique regional sector strengths or vulnerabilities.
- **Outlier Analysis:** Investigating the outliers within sectors or continents to understand the drivers behind extreme changes in multiples.


```python
# Temporal Trends - Calculating yearly mean change in EV/EBITDA multiple by sector and continent
company_level_metrics['year'] = company_level_metrics['record_date'].dt.year
yearly_change_by_sector = company_level_metrics.groupby(['sector_level_1', 'year'])['change_in_ebitda_multiple'].mean().unstack().T
yearly_change_by_continent = company_level_metrics.groupby(['continent', 'year'])['change_in_ebitda_multiple'].mean().unstack().T

# Visualization of Yearly Mean Change in EV/EBITDA Multiple by Sector

# Create figure and subplots with different sizes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), gridspec_kw={'width_ratios': [3, 2]})

# Plot 1: Yearly Mean Change in EV/EBITDA Multiple by Sector
sns.heatmap(yearly_change_by_sector, cmap='RdBu', annot=True, fmt=".2f", ax=ax1)
ax1.set_title('Yearly Mean Change in EV/EBITDA Multiple by Sector')
ax1.set_xlabel('Sector')
ax1.set_ylabel('Year')
ax1.tick_params(axis='x', rotation=45)

# Plot 2: Yearly Mean Change in EV/EBITDA Multiple by Continent
sns.heatmap(yearly_change_by_continent, cmap='RdBu', annot=True, fmt=".2f", ax=ax2)
ax2.set_title('Yearly Mean Change in EV/EBITDA Multiple by Continent')
ax2.set_xlabel('Continent')
ax2.set_ylabel('Year')
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

```

From the plot above on the metrics on yearly mean change data, we can derive:

### Yearly Mean Change in EV/EBITDA Multiple by Sector:
- **Sector Trends:** The heatmap shows varying patterns of change across different sectors over the years. Certain sectors exhibit periods of significant positive or negative changes, reflecting the dynamic nature of their market valuations.
- **Temporal Fluctuations:** The yearly fluctuations within sectors can be indicative of external economic factors, sector-specific developments, or changes in market sentiment. For example, periods of substantial positive change may align with technological breakthroughs, regulatory changes, or shifts in consumer behavior.

### Yearly Mean Change in EV/EBITDA Multiple by Continent:
- **Geographical Variability:** Similar to the sector analysis, the heatmap for continents reveals variations in the yearly mean change of multiples. This illustrates how different regions may respond to global economic conditions, regional events, or market trends.
- **Comparison Across Continents:** The visualization allows for a direct comparison of trends across continents, highlighting years where certain regions experienced more pronounced valuation changes. This comparison can be useful for understanding regional economic cycles or the impact of global market conditions.
